## 4. Задание (в программе)
1) Решите систему уравнений:

y = x^2 – 1

exp(x) + x*(1 – y) = 1

```python
from scipy.optimize import fsolve
```

```python
plt.figure(figsize=[5,5])
x = np.linspace(-6, 6, 201)
plt.plot(x, x**2 - 1)
plt.plot(x, (np.exp(x) + x - 1) / x)
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
def equations(p):
    x, y = p                                         ## TWO sep values will become x,y from p
    return(y - x**2 + 1, np.exp(x) + x - x * y - 1)  ## both eq are equal to zero
```

```python
x1, y1 = fsolve(equations, (-2,2))
print(f'кординаты корня:  X = {round(x1,4)}, Y= {round(y1,4)}')
```

```python
x1, y1 = fsolve(equations, (2,4))
print(f'кординаты корня:  X = {round(x1,4)}, Y= {round(y1,4)}')
```

```python
x1, y1 = fsolve(equations, (4,4))
print(f'кординаты корня:  X = {round(x1,4)}, Y= {round(y1,4)}')
```
